```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
x = np.linspace(0,(1), 100)

def exp(x):
    return np.exp(x)
    
```

```python
y = exp(x)

def plot(x,y, flag_save=True, xlabel='Time [milliseconds]', ylabel='Awesomeness', 
         lcolor='blue', fs=14, fname='plot,png'):
    f, ax = plt.subplots(1,1,figsize=(4,4))
    ax.plot(x,y,color=lcolor,linewidth=1.5)
    ax.set_xlabel(xlabel,fontsize=fs)
    ax.set_ylabel(ylabel,fontsize=fs)
    if(flag_save):
        plt.savefig(fname, format="pdf", bbox_inches='tight',dpi=400)

def main():
    x = np.linspace(0,1,10)
    y = x**2
    plot(x,y)

if __name__=='__main__':
    main()
```
